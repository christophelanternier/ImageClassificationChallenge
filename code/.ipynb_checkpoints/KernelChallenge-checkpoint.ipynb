## SVM implementation

```python
y = np.array([1, -1, 1, -1])
```

```python
X = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
```

```python
def transform_svm_dual( C, X, y ):
    
    (d, n) = X.shape
    M = (y*X);
    Q = M.T.dot(M)
    p = -np.ones(n)
    b = np.zeros(2*n)
    b[:n] = C
    A = np.zeros((2*n , n))
    A[:n , :n] = np.eye(n)
    A[n:, :n] = -np.eye(n)

    return Q,p,A,b
```

```python
def f1(x, t):
    B=0
    for i in range(A.shape[0]):
        B = B -np.log(b[i]-A[i,:].dot(x))

    return t*(1/2)*(x.T).dot(Q.dot(x))+t*(p).dot(x)+B

def grad_f1(x, t):
    c = np.zeros(A.shape[0]);
    for i in range(len(c)):
        c[i] = 1/(b[i]-A[i,:].dot(x))

    return t*Q.T.dot(x) + t*p + A.T.dot(c)


def hess_f1(x, t):

    c = np.zeros(A.shape[0]);
    for i in range(len(c)):
        c[i] = 1/(b[i]-A[i,:].dot(x))

    M1 = np.zeros((A.shape[1], A.shape[1]))
    for i in range(len(c)):
        M1 = M1 + A[[i], :].T.dot(A[[i], :])/(b[i]-A[i,:].dot(x))

    return t*Q + M1
```

```python
def Newton(t,x, grad_f1, hess_f1, f1):
    beta = 1/2
    alpha = 1/4

    delta_x = -np.linalg.inv(hess_f1(x,t)).dot(grad_f1(x, t))
    lambda1 = np.sqrt(grad_f1(x, t).T.dot(np.linalg.inv(hess_f1(x,t)).dot(grad_f1(x, t))))

    #Backtracking line search
    step = 1

    while (f1(x+step*delta_x, t) >= f1(x, t)+alpha*step*grad_f1(x, t).T.dot(delta_x)) or ~np.isreal(f1(x+step*delta_x, t)):
        step = beta*step

    x_new = x+step*delta_x
    gap = lambda1**2/2
    print x_new
    
    return x_new, gap
```

```python
def centering_step(Q,p,A,b,x,t,tol):

    gap = np.inf
    #x_seq = []
    
    while gap > tol:
        [x_new, gap] = Newton(t,x, grad_f1, hess_f1, f1)
        #x_seq = x_seq.append(x_new)
        x = x_new

    return x
```

```python
def barr_method(Q,p,A,b,x_0,mu,tol):

    x = x_0
    #x_seq_total = []
    m = A.shape[0]
    t=1

    while m/t >= tol:

        x = centering_step(Q,p,A,b,x,t,tol)

        #x_seq_total = x_seq_total.append(x_seq)
        #x = x_seq[-1]
        t = mu*t
       

    #x_sol = x_seq_total[-1][-1]
    return x

    #return x_sol, x_seq_total
```
